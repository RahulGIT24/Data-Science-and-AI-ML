## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup Plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. Plot the data
ax.plot(x,y)

# 4. Customize Plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save and show
fig.savefig("../Images/sample-plot.png")
```

# Two options for subplots

```python
# Option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,5))
# Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(100));
```

```python
# Subplots Option 2
fig, ax = plt.subplots(2,2,figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(100),np.random.random(100))
ax[1,0].hist(np.random.random(100));
ax[1,1].bar(nut_butter_prices.keys(), nut_butter_prices.values())
```

### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly use pyplot method
* When Plotting something advanced use OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,6))
plt.show()
```

```python
# OO Method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            title="Over 50 heart data",
            ax=ax);
```

```python
## oo Method from scratch
fig,ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter  = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
# Customize
ax.set(title="Heart Disease and Cholestrol Level",xlabel="Age", ylabel="Cholestrol")
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
# Add a horizontal line
ax.axhline(y=over_50["chol"].mean(), linestyle="--")
plt.show()
```

```python
over_50.head()
```

```python
# Subplot of Chol, age thalach
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,6), sharex=True)

# Plot the data
scatter  = axes[0].scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize axes[0,0]
axes[0].set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")

# Add a legend
axes[0].legend(*scatter.legend_elements(), title="Target")

# Add a mean line
axes[0].axhline(y=over_50["chol"].mean(),
                 linestyle="--")

# Add data to axes[0,1]
scatter = axes[1].scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
axes[1].set(title="Heart Disease and Thalach", xlabel="Age", ylabel="Thalach")
axes[1].legend(*scatter.legend_elements(),title="Target")
axes[1].axhline(y=over_50["thalach"].mean(), linestyle="--")

# Added title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```

```python
# Customizing the x and y limitations
# Subplot of Chol, age thalach
plt.style.use("classic")
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,6), sharex=True)

# Plot the data
scatter  = axes[0].scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"],cmap='summer')

# Customize axes[0,0]
axes[0].set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")

# Add a legend
axes[0].legend(*scatter.legend_elements(), title="Target")
# axes[0].set_xlim([50,80])

# Add a mean line
axes[0].axhline(y=over_50["chol"].mean(),
                 linestyle="--")

# Add data to axes[0,1]
scatter = axes[1].scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap='winter')
axes[1].set(title="Heart Disease and Thalach", xlabel="Age", ylabel="Thalach")
axes[1].legend(*scatter.legend_elements(),title="Target")
axes[1].axhline(y=over_50["thalach"].mean(), linestyle="--")

# Added title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
plt.subplots_adjust(hspace=0.4)

plt.show()
```
